```python
import random
n=int(input("enter the number times you want to play :"))

r='rock'
s='scissors'
p='paper'
c=[r,s,p]
pl_s=0
cmp_s=0
tie_count=0
i=1
comp=random.choice(c)
while i<=n:
    pl=input("enter the your choice from the ROCK SCISSORS PAPER : ")
    if(pl==comp):
        tie_count+=1
        print("it is the tie ")
        print("the computer had also choosen the same so it is tie")
    elif(pl==r):
        if(comp==s):
            print ("***you are the winner****")
            pl_s+=1
            print("the computer has choosen the scissors ")
            print("since the rock can break the scissors so there is no more scissors ")
        else:
            print("**the computer is the winner **")
            print("since the computer has chosen the ***paper**so:")
            print("the paper can cover the rock then there is no rock can seen  ")
            
            cmp_s+=1
    elif(pl==p):
        if(comp==r):
            print ("***you are the winner****")
            print("the computer had choosen the rock")
            print("since the paper can cover the rock then there is no rock can seen  ")
            pl_s+=1
        else:
            print("**the computer is the winner **")
            print("**since computer has choosen **sciisors** so:")
            print("** the scisssors can cut the paper then there is no paper ")
            cmp_s+=1
    elif(pl==s):
        if(comp==p):
            print ("***you are the winner****")
            print(" the computer has choosen the paper ")
            print("since the scisssors can cut the paper then there is no paper ")
            pl_s+=1
        else:
            print("**the computer is the winner **")
            print("****the computer choosen ***rock*** so:")
            print("*** the rock can break the scissors so there is no more scissors** ")
            cmp_s+=1
    
    i+=1

if(n<i):  
    print("the user score is {}".format(pl_s))
    print("the computer score is {}".format(cmp_s))
    print("the tie count is {} ".format(tie_count))
if(pl_s>cmp_s):
    print("the winner is user")
elif(cmp_s==pl_s):
    print("equal scores ")
else:
    print("the computer is the winner ")
```
